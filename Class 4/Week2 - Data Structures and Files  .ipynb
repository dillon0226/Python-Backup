```python
# String Slicing - returning subsrings of strings  
# A slice INCLUDES the character at the start index but EXCLUDES the character at the end index

weather = "thunderstorm"
```

```python
# return 'storm' from 'thunderstorm'
# 's' is at index 7 

weather[7:]
```

```python
# return 'under' from 'thunderstorm' 
# 'u' is at index 2 and 'r' is at index 6

weather[2:7]
```

```python
# we miss the value if we end the slice at the characters index 

weather[2:6]
```

```python
# full string         [:]
# start of string    [:x]
# end of string      [x:]

print(weather[:])
print(weather[:4])
print(weather[4:])
```

```python
# we can use concatentation to join the last two results and create
# the original string 

weather[:4] + weather[4:]
```

```python
# if the index is negative, Python counts back from 
# the end of the string instead of the beginning 

print(len(weather))

# go back 12 characters to end of string 
print(weather[-12:])

# beginning of string up to the last 3 characters  
print(weather[:-3])
```

```python
# if the index is beyond the end of the string
# returns an empty string 

print(weather[50:])
print(weather[:-30])
```

```python
# use slicing and joining to format a US phone number 
# (xxx)-xxx-xxxx

phonenumber = "4123456789"

# first 3 digits are area code 
area_code = "(" + phonenumber[0:3] + ")"

print(area_code)
```

```python
# next 3 digists are the exchange 

exchange_code = phonenumber[3:6]
print(exchange_code)
```

```python
# last 4 digits are the line number 
# can obtain with negative slicing 
# go back 4 elements to end of string

line = phonenumber[-4:]
line
```

```python
# join them together 

# using + 
print(area_code + "-" + exchange_code + "-" + line)

# using .join()
number_list = [area_code,exchange_code,line]
print("-".join(number_list))
```

```python
# use it in a Function => resuable block code 

def format_phone(phonenumber):
    # first 3 digits are area code 
    area_code = "(" + phonenumber[0:3] + ")"

    # next 3 digists are the exchange 
    exchange_code = phonenumber[3:6]

    # last 4 digits are the line number 
    line = phonenumber[-4:]

    # using .join()
    number_list = [area_code,exchange_code,line]
    return "-".join(number_list)

print(format_phone("4123456789"))
```

```python
# String Formatting 
# Syntax:
#            Call the .format() method on a string, passing the values to be inserted as arguments.
# Placeholders:
#            Use curly braces {} within the string as placeholders.
# Positional Arguments:
#            Placeholders can be empty {} or contain numbers {0}, {1} corresponding to argument positions.
# Named Arguments:
#            Placeholders can contain names {name} corresponding to keyword arguments passed to .format().
# Format Specifiers:
#            Format specifiers can be used within the curly braces.
```

```python
# String Formatting 
# Alignment 
#            < left, > right, ^ center 
# Precision 
#            .<number> 
#             floats - specifies number of decimal places 
#             strings - limits the number of characters 
# Type 
#             d = integer 
#             f = float 
#             s = string
#             b = binary 
#             % = percentage 
```

```python
# format a float as an integer value 
# {:f}

"{:.0f}".format(10.5)
```

```python
# format a string with x number of characters 
# {:.xs]

python = "Python"

"{:.2s}".format(python)
```

```python
# align python 6 spaces to the right 
# {:>6s}

"{:>6s}".format("Py")

```

```python
# align python 6 spaces to the left 
# {:<6s}

"{:<6s}".format("Py")
```

```python
# align right AND reduce characters 

"{:>6.2}".format(python)
```

```python
# center string by x spaces 

"{:^6s}".format("Py")
```

```python
# Dictionary - mutable, curly brackets {}, key/value pairs 

# create a dictionary based on counts of file types 

file_counts = {"jpg":10, "txt":14, "csv":3, "py":13}
print(file_counts)
```

```python
# update() dictionary 

# modify value of py counts 
# ** REPLACES, does not append 
file_counts.update({"py":21})
print(file_counts)

# add new key/value pair to dictionary 
file_counts.update({"pdf":32})
print(file_counts)
```

```python
# append value to existing key 
# ERROR 
file_counts["py"].append(13)
```

```python
# append value to existing key 
# convert to data type of values to list 

# Convert each value to a list containing that value
file_counts = {key: [value] for key, value in file_counts.items()}

file_counts["py"].append(13)
print(file_counts)
```

```python
# remove index 1 of py file type values 
file_counts["py"].pop(1)
```

```python
# remove txt value 
file_counts["txt"].remove(14)
```

```python
# remove txt key and value pair 
file_counts.pop("txt")
```

```python
# open file 
file = open("sports.txt")

# read file line by line and return as string 
print(file.readline())
print(file.readline())


# read rest of full file and returns as string
print(file.read())

# when files are opened in python,
# they should always be closed after use
# this frees up system resources, 
# save all data, prevent data corruption, and avoid errors

file.close()
```

```python
# execute a file open, file read, and auto file close  
# within one code block 

with open("sports.txt") as file:
    print(file.read())
```

```python
# iterate through files 

with open("sports.txt") as file:
    for line in file:
        print(line.upper())

# Each line in a text file already ends with a newline character (\n)
# line includes the newline at the end (e.g., "soccer\n").
# print() adds another newline after printing by default.
# hence the double spacing 
```

```python
# iterate through files and remove newlines 

with open("sports.txt") as file:
    for line in file:
        print(line.strip().upper())
```

```python
# write to a file 

with open("car_brands.txt", "w") as file:
    file.write("Toyot\nHonda\nFord\nJeep\nVolkswagen\nMercedes Benz")
```
