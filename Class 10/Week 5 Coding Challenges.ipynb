```python
students = {
    "Name": ["Alice", "Bob", "Charlie", "Diana", "Ethan"],
    "Year": [2, 1, 4, 4, 3],
    "GPA": [3.5, 3.8, 3.2, 3.9, 3.4]
}

df = pd.DataFrame(students)
print("Original DataFrame:")
print(df)

# Filter by GPA > 3.5
print("\n\nStudents with GPA > 3.5:")
high_gpa = df[df['GPA'] > 3.5]
print(high_gpa)

# Add column for Graduating students (if year == 4)
df['Graduating'] = df['Year'] == 4
print("\n\nDataFrame with Graduating column:")
print(df)

# Alternative: Use 'Yes'/'No' instead of True/False
df['Graduating'] = df['Year'].apply(lambda x: 'Yes' if x == 4 else 'No')
print("\n\nDataFrame with Graduating column (Yes/No):")
print(df)
```

```python
# Challenge 2 
# Merging vs Concatenating DataFrames

# Three small DataFrames with student info
df1 = pd.DataFrame({
    "StudentID": [1, 2, 3],
    "Name": ["Alice", "Bob", "Charlie"],
    "Year": [2, 1, 4]
})

df2 = pd.DataFrame({
    "StudentID": [2, 3, 4],
    "GPA": [3.8, 3.2, 3.9]
})

df3 = pd.DataFrame({
    "StudentID": [5, 6],
    "Name": ["Diana", "Ethan"],
    "Year": [4, 3]
})

print("DataFrame 1:\n", df1, "\n")
print("DataFrame 2:\n", df2, "\n")
print("DataFrame 3:\n", df3, "\n")
```

```python
# 2.1 Concatenate df1 and df3 vertically (stack rows)

# Three small DataFrames with student info
df1 = pd.DataFrame({
    "StudentID": [1, 2, 3],
    "Name": ["Alice", "Bob", "Charlie"],
    "Year": [2, 1, 4]
})

df2 = pd.DataFrame({
    "StudentID": [2, 3, 4],
    "GPA": [3.8, 3.2, 3.9]
})

df3 = pd.DataFrame({
    "StudentID": [5, 6],
    "Name": ["Diana", "Ethan"],
    "Year": [4, 3]
})

print("DataFrame 1:\n", df1, "\n")
print("DataFrame 2:\n", df2, "\n")
print("DataFrame 3:\n", df3, "\n")

# Concatenate df1 and df3 vertically
df_combined = pd.concat([df1, df3], axis = 0)

print("Combined DataFrame (df1 + df3):")
print(df_combined)
```

```python
# 2.2 Concatenate df1 and df3 horizontally (side by side)

# Three small DataFrames with student info
df1 = pd.DataFrame({
    "StudentID": [1, 2, 3],
    "Name": ["Alice", "Bob", "Charlie"],
    "Year": [2, 1, 4]
})

df2 = pd.DataFrame({
    "StudentID": [2, 3, 4],
    "GPA": [3.8, 3.2, 3.9]
})

df3 = pd.DataFrame({
    "StudentID": [5, 6],
    "Name": ["Diana", "Ethan"],
    "Year": [4, 3]
})

print("DataFrame 1:\n", df1, "\n")
print("DataFrame 2:\n", df2, "\n")
print("DataFrame 3:\n", df3, "\n")

# Concatenate df1 and df3 horizontally
df_combined = pd.concat([df1, df3], axis = 1)

print("Combined DataFrame (df1 + df3):")
print(df_combined)
```

```python
# 2.3 Merge df1 and df2 on StudentID (inner join)

# Three small DataFrames with student info
df1 = pd.DataFrame({
    "StudentID": [1, 2, 3],
    "Name": ["Alice", "Bob", "Charlie"],
    "Year": [2, 1, 4]
})

df2 = pd.DataFrame({
    "StudentID": [2, 3, 4],
    "GPA": [3.8, 3.2, 3.9]
})

df3 = pd.DataFrame({
    "StudentID": [5, 6],
    "Name": ["Diana", "Ethan"],
    "Year": [4, 3]
})

df_merged = pd.merge(df1, df2, on='StudentID', how='inner')

print("Merged DataFrame (inner join on StudentID):")
print(df_merged)
```

```python
# Three small DataFrames with student info
df1 = pd.DataFrame({
    "StudentID": [1, 2, 3],
    "Name": ["Alice", "Bob", "Charlie"],
    "Year": [2, 1, 4]
})

df2 = pd.DataFrame({
    "StudentID": [2, 3, 4],
    "GPA": [3.8, 3.2, 3.9]
})

df3 = pd.DataFrame({
    "StudentID": [5, 6],
    "Name": ["Diana", "Ethan"],
    "Year": [4, 3]
})

df_merged = pd.merge(df1, df2, on='StudentID', how='outer')

print("Merged DataFrame (outer join on StudentID):")
print(df_merged)
```

```python
# Challenge 3
# use the map() function to create a new "Distinguished Honor Roll" column 
# to map GPA values >= 3.8 to true, else false 

# .map() expects either:
# a function (to apply element‑wise)
# a dictionary (for lookup)
# or another DataFrame series (for alignment)

# when you want custom logic (like if val > 5 then …), you need a function
# instead of defining a full function with def, 
# you can use a lambda — a quick function written inline.
```

```python
# Create students DataFrame
students = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve', 'Frank'],
    'Age': [20, 21, 19, 22, 20, 23],
    'GPA': [3.8, 3.2, 3.9, 3.4, 3.7, 3.1],
    'Year': [2, 3, 1, 4, 3, 4]
}

df = pd.DataFrame(students)
print("Original DataFrame:")
print(df)

# Use map() with a lambda function to create Distinguished Honor Roll column
df['Distinguished Honor Roll'] = df['GPA'].map(lambda x: True if x >= 3.8 else False)

print("\nDataFrame with Distinguished Honor Roll column:")
print(df)
```

```python
import matplotlib.pyplot as plt
```

```python
# Challenge 1
# Create Line chart to show GPA vs Year
```

```python
# Create students DataFrame
students = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve', 'Frank'],
    'Age': [20, 21, 19, 22, 20, 23],
    'GPA': [3.8, 3.2, 3.9, 3.4, 3.7, 3.1],
    'Year': [2, 3, 1, 4, 3, 4]
}

df = pd.DataFrame(students)

# Create scatter plot with line
plt.figure(figsize=(10, 6))
plt.plot(df['Year'], df['GPA'], marker='o', linestyle='-', linewidth=2, markersize=10, color='blue')

# Add student names as labels
for i, name in enumerate(df['Name']):
    plt.annotate(name, (df['Year'][i], df['GPA'][i]),
                textcoords="offset points", xytext=(0,10), ha='center')

# Add labels and title
plt.xlabel('Year', fontsize=12)
plt.ylabel('GPA', fontsize=12)
plt.title('GPA vs Year (by Student)', fontsize=14, fontweight='bold')

# Add grid
plt.grid(True, alpha=0.3)

# Set axis limits
plt.xlim(0, 5)
plt.ylim(2.5, 4.0)

plt.tight_layout()
plt.show()
```

```python
# Challenge 2 
# Create a Histogram of GPA distribution 
```

```python
# Create students DataFrame
students = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve', 'Frank'],
    'Age': [20, 21, 19, 22, 20, 23],
    'GPA': [3.8, 3.2, 3.9, 3.4, 3.7, 3.1],
    'Year': [2, 3, 1, 4, 3, 4]
}

df = pd.DataFrame(students)

# Create histogram for GPA
plt.figure(figsize=(10, 6))
plt.hist(df['GPA'], bins=5, color='skyblue', edgecolor='black', alpha=0.7)

# Add labels and title
plt.xlabel('GPA', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.title('Distribution of GPA', fontsize=14, fontweight='bold')

# Add grid
plt.grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.show()
```

```python
# Challenge 2
# Boxplot GPA by Graduating values 
# split GPA distribution by graduating status
```

```python
# Create students DataFrame
students = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve', 'Frank'],
    'Age': [20, 21, 19, 22, 20, 23],
    'GPA': [3.8, 3.2, 3.9, 3.4, 3.7, 3.1],
    'Year': [2, 3, 1, 4, 3, 4]
}

df = pd.DataFrame(students)

# Add Graduating column
df['Graduating'] = df['Year'].apply(lambda x: 'Yes' if x == 4 else 'No')

print("DataFrame:")
print(df)

# Create boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Graduating', y='GPA', palette='Set2')

# Add labels and title
plt.xlabel('Graduating Status', fontsize=12)
plt.ylabel('GPA', fontsize=12)
plt.title('GPA Distribution by Graduating Status', fontsize=14, fontweight='bold')

# Add grid
plt.grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.show()
```
