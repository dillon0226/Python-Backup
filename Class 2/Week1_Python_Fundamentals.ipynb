```python
# 2 
# Use variables to store values 
# Use Logical and Relational Operators 

# Write a conditional block to determine a student's grade on a test based on their score as a variable 
# and this provided grading scale 

# A for 90 and above
# B for 8089
# C for 7079
# D for 6069
# F for below 60
```

```python
student_score = 100

if student_score < 0 or student_score > 100:
    print ("Invalid score")
elif student_score >= 90:
    grade = "A"
    print("Your grade is: " + grade)
elif student_score >= 80:
    grade = "B"
    print("Your grade is: " + grade)
elif student_score >= 70:
    grade = "C"
    print("Your grade is: " + grade)
elif student_score >= 60:
    grade = "D"
    print("Your grade is: " + grade)
else:
    grade = "F"
    print("Your grade is: " + grade)
```

```python
#Create a conditional block to test and output the type of value
#string, int, float, boolean, none
#ID an edge case

value = False
# As written, this will print an integer because False is 0 and we check for integer before boolean.
if isinstance(value, str):
    print("The value is a string.")
elif isinstance(value, int):
    print("The value is an integer.")
elif isinstance(value, float):
    print("The value is a float.")
elif isinstance(value, bool):
    print("The value is a boolean.")
elif value == None:
    print("The value is none.")
else:
    print("Unknown type.")
```
